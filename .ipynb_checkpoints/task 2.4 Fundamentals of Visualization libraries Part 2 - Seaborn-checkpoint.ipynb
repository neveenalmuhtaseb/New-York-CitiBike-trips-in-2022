## What are the busiest months and is there a weather component at play?

```python
df['date'] = pd.to_datetime(df['date'], format = '%Y-%m-%d')
```

```python
# Groupby the count of the trips 

df_group = pd.DataFrame(df.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
df_group.head()
```

```python
df_group.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)
df = df.merge(df_group, on = "date", how = 'outer', indicator = 'merge_flag')
print(df['merge_flag'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['date'] == '2018-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['date'] == '2018-01-02'].shape) # Second check 

df_temp = df.set_index('date')

print(df_temp.columns)
```

```python
df_temp.columns
```

```python
fig,ax = plt.subplots(figsize=(10, 5))
sns.lineplot(data = df_temp['bike_rides_daily'], color = "b")
ax.set_xlabel("Year 2022", fontsize = 14) # set x-axis label
ax.set_ylabel("Bike rides daily",color = "navy", fontsize = 14) # set y-axis label

ax2 = ax.twinx()
sns.lineplot(data = df_temp['avgTemp'], color="r", ax=ax2)
ax2.set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.title('Temperature and trips in 2022', fontsize = 18)
# Save the figure to a file 
plt.savefig("Temperature and trips", bbox_inches='tight')
plt.show()
```

Bike usage peaks between May and October, likely driven by better weather conditions that make cycling more appealing.

## What is the distribution of trip durations?

```python
df.dtypes
```

```python
# First, convert to datetime if not already
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

# Create a new 'tripduration' column in minutes (or seconds if you prefer)
df['trip_duration'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60
```

```python
from scipy.stats import gaussian_kde

# Histogram + KDE
fig, ax = plt.subplots(figsize=(8, 5))

# Filter out extreme trip durations (e.g., >120 mins) if desired
filtered = df[df['trip_duration'] <= 120]['trip_duration']

counts, bins, patches = ax.hist(filtered, bins=30, color='skyblue', alpha=0.7, density=True, label='Histogram')

kde = gaussian_kde(filtered)
x_vals = np.linspace(filtered.min(), filtered.max(), 1000)
ax.plot(x_vals, kde(x_vals), color='darkblue', lw=2, label='KDE Curve')

ax.set_title('Trip Duration Histogram with KDE Curve')
ax.set_xlabel('Trip Duration (minutes)')
ax.set_ylabel('Density')
ax.legend()

plt.show()
```

```python
# Remove outliers above 99th percentile
threshold = df['trip_duration'].quantile(0.99)
df_filtered = df[df['trip_duration'] < threshold]

# Plot boxplot using the filtered data (without outliers)
sns.boxplot(data=df_filtered, y='trip_duration')
plt.title("Distribution of Trip Duration (Filtered below 99th percentile)")
plt.ylabel("Duration (seconds)")
# Save the figure to a file 
plt.savefig("Distribution of Trip Duration", bbox_inches='tight')
plt.show()
```

The most frequent trip durations fall within the 5 to 20-minute range. Trips lasting less than 5 minutes (the first quartile) are the next most common. Longer trips, ranging from 25 to 50 minutes, occur less frequently.

## Is there a difference in trip duration between customer types?

```python
grid = sns.FacetGrid(df_filtered, col = "member_casual")
grid.map(sns.histplot, "trip_duration", bins = 15)
# Save the figure to a file 
plt.savefig("Trip Duration vs customer types", bbox_inches='tight')
plt.show()
```

 There is no significant difference in trip duration between customer types; both groups typically use the bikes for 5 to 20 minutes. However, as mentioned earlier, members tend to use the bikes more frequently than casual customers.
